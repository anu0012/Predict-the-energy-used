```python
feature_names = [x for x in train.columns if x not in ['Observation','Energy']]
target = train['Energy']
```

```python
# Current best
from catboost import CatBoostRegressor

model_catboost = CatBoostRegressor(eval_metric='RMSE',learning_rate=0.3,verbose=True, iterations=500,depth=10)

model_catboost.fit(train[feature_names], target)

pred = model_catboost.predict(test[feature_names])

pred[:10]
```

```python
## make submission
sub = pd.DataFrame()
sub['Observation'] = test['Observation']
sub['Energy'] = pred
sub.to_csv('result.csv', index=False)
```
